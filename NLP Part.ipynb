```python
def text_to_embed(text: str, tokenizer, model):
    text=[text]
    inputs = tokenizer(text, padding=True, truncation=True, max_length=512, return_tensors="pt").to(device)  # Îç∞Ïù¥ÌÑ∞Î•º GPUÎ°ú Ïù¥Îèô
    
    with torch.no_grad():
        embeddings = model(**inputs, return_dict=True).last_hidden_state[:, 0]
        embeddings = embeddings / embeddings.norm(dim=1, keepdim=True)
    
    return embeddings[0]
```

### ÏòàÏãú ÎèåÎ¶¨Í∏∞

```python
user_cache={} #key: user_id, value: embeded vector
item_cache={} #key: item_id, value: embeded vector
sim_cache={}   #key: (user_id, item_id), value: cosine similarity
```

```python
import json

# JSON ÌååÏùºÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
with open('user_cache.json', 'w') as json_file:
    loaded_data = json.load(json_file)
    
with open('item_cache.json', 'w') as json_file:
    loaded_data = json.load(json_file)
    
with open('item_cache.json', 'w') as json_file:
    loaded_data = json.load(json_file)
```

```python
import json

def calculate_cosine(mode: str, n: int, ngcf: dict, reviews: pd.DataFrame, item: pd.DataFrame): #mode='review', 'LLM' #ngcfÎ°ú Ï∂îÏ∂úÌïòÎäî item Í∞úÏàò
    #user_cache, item_cache, sim_cache -> comprehension
    users_list=list(ngcf.keys())
    
    #embedding and save user text
    for user in users_list:
        if user not in user_cache.keys():
            user_cache[user]=text_to_embed(reviews[reviews['user_id']==user][mode].iloc[0])
            
        for item_rank in range(n):
            #embedding and save item text
            item_id, score = ngcf[user][item_rank]
            if item_id not in item_cache.keys():
                item_cache[item_id]=text_to_embed(item[item['item_id']==item_id]['item_text'].iloc[0])
            
            #calculate cosine similarity
            if (user, item_id) not in sim_cache.keys():
                sim_cache[(user, item_id)]=user_cache[user]@item_cache[item_id]
    
    #save cache
    with open('user_cache.json', 'w') as json_file:
        json.dump(user_cache, json_file)
    with open('item_cache.json', 'w') as json_file:
        json.dump(item_cache, json_file)
    with open('sim_cache.json', 'w') as json_file:
        json.dump(sim_cache, json_file)
            
    return sim_cache
```
